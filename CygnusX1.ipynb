```python
from ixpeobssim import IXPEOBSSIM_CALDB
#from ixpeobssim.irf.caldb import irf_file_name
# Create the polarization cubes.

kwargs = dict( algorithm ='PCUBE',ebinning=ENERGY_BINNING , overwrite = OVERWRITE,irfname = 'ixpe:obssim:v12')
src_pcube_file_list = pipeline.xpbin (*src_file_list, ** kwargs,mc= False)
bkg_pcube_file_list = pipeline.xpbin (*bkg_file_list, ** kwargs,mc = False)

# Read back the polarization cubes and perform the background subtraction.
src_pcube = xBinnedPolarizationCube . from_file_list (src_pcube_file_list )
bkg_pcube = xBinnedPolarizationCube . from_file_list (bkg_pcube_file_list )
bkg_pcube *= src_pcube .backscal () / bkg_pcube .backscal ()
src_pcube -= bkg_pcube
```

```python
import matplotlib.pyplot as plt

# normalized stokes parameters
Q_I = src_pcube.Q/src_pcube.I
U_I = src_pcube.U/src_pcube.I
u_i = numpy.linspace(-5,5,10)
q_i = numpy.linspace(-5,5,10)
polarization_degrees = [1, 3, 5, 7, 9]
polarization_angles = [0, 45, 90,-75,-60,-45,-20,-15]
def plot_polarization_circle(ax, degree):
    circle = plt.Circle((0, 0), degree, fill=False, edgecolor='black')
    ax.add_patch(circle)
    if degree <= 5:
      ax.text(degree, 0, f'{degree*10}°', ha='center', va='center', color='black', fontsize=10)
fig, ax = plt.subplots()
for degree in polarization_degrees:
    plot_polarization_circle(ax, degree)
ax.set_aspect('equal')
print(Q_I,U_I)
ax.set_xlim(-6, 6)
ax.set_ylim(-6, 6)
color = ['red','blue','green']
label = ['2-4 kev','4-6 kev','6-8 kev']
for i in range(len(Q_I)):
  plt.scatter(Q_I[i],U_I[i],color = color[i],label = label[i])
plt.legend()
plt.xlabel('normalized Q/I')
plt.ylabel('normalized U/I')
plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, sharex=True)

for i in range(len(ENERGY_BINNING)-1):
    ax[0].scatter(ENERGY_BINNING[i],src_pcube.PD[i])
    ax[0].errorbar(ENERGY_BINNING[i], src_pcube.PD[i], yerr=src_pcube.PD_ERR[i], label=f'{ENERGY_BINNING[i]}-{ENERGY_BINNING[i+1]} keV')
    ax[1].scatter(ENERGY_BINNING[i],src_pcube.PA[i])
    ax[1].errorbar(ENERGY_BINNING[i], src_pcube.PA[i], yerr=src_pcube.PA_ERR[i], label=f'{ENERGY_BINNING[i]}-{ENERGY_BINNING[i+1]} keV')

ax[0].set_xlim(1,8)
ax[0].set_ylabel('Polarization degree')
ax[1].set_ylabel('Polarization angle (rad)')
ax[1].set_xlabel('Energy')
ax[0].legend()
ax[1].legend()

plt.show()
```
